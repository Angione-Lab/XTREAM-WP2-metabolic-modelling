```python
import pandas as pd
import os
from cobra.io import read_sbml_model
from cobra.flux_analysis import flux_variability_analysis
from cobra.flux_analysis.loopless import add_loopless, loopless_solution
from cobra.flux_analysis import pfba
import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
```

### Individual model flux estimation

```python
met_files = ['../models\\modelGb1_bin16.xml',
 '../models\\modelGb1_bin17.xml',
 '../models\\modelGb1_bin20.xml',
 '../models\\modelGb1_bin43.xml',
 '../models\\modelGb2_bin20.xml',
 '../models\\modelGb2_bin21.xml',
 '../models\\modelGb4_bin17.xml',
 '../models\\modelGb7_bin10.xml',
 '../models\\modelGb7_bin15.xml',
 '../models\\modelGb7_bin28.xml',
 '../models\\modelGb7_bin5.xml',
 '../models\\modelGb7_bin9.xml',
 '../models\\modelGb8_bin5.xml',
 '../models\\modelGb8_bin50.xml']
```

### FVA

```python
minFluxes = pd.DataFrame()
maxFluxes = pd.DataFrame()
for met in tqdm.tqdm(met_files):
    model = read_sbml_model(met)
    FVA  = flux_variability_analysis(model, model.reactions, fraction_of_optimum=0.9)
    minFluxes[os.path.splitext(os.path.basename(met))[0]] = FVA['minimum']
    maxFluxes[os.path.splitext(os.path.basename(met))[0]] = FVA['maximum']

maxFluxes.to_csv("../results/maxfluxes.csv")
minFluxes.to_csv("../results/minfluxes.csv")
```

### pFBA

```python
Fluxes = pd.DataFrame()
for met in tqdm.tqdm(met_files):
    model = read_sbml_model(met)
    solution = pfba(model)
    Fluxes[os.path.splitext(os.path.basename(met))[0]] = solution.fluxes

Fluxes.to_csv("../results/pFBA_fluxes.csv")
```
